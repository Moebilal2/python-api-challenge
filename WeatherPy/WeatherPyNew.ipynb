```python
import requests
import json
from api_keys import weather_api_key
from pprint import pprint
import time
from scipy.stats import linregress
# Incorporated citipy to determine city based on latitude and longitude
from citipy import citipy
import numpy as np
import random
import time
import pandas as pd
import matplotlib.pyplot as plt
outputfile="output_data/cities.csv"
```

```python
latrange=(-100,100)
lonrange=(-150,150)
latlon=[]
cities=[]
lat=np.random.uniform(latrange[0],latrange[1],size=1600)
lon=np.random.uniform(lonrange[0],lonrange[1],size=1600)
latlon=zip(lat,lon)

for latlon in latlon:
    city=citipy.nearest_city(latlon[0],latlon[1]).city_name
    if city not in cities:
        cities.append(city)
```

```python
len(cities)
```

```python
#https://api.openweathermap.org/data/2.5/weather?lat={lat}&lon={lon}&appid={API key}
url = "http://api.openweathermap.org/data/2.5/weather?units=Imperial&APPID=" + weather_api_key
city_data=[]
# latitude=[]
# longitude=[]
# maxtemp=[]
# mintemp=[]
# hum=[]
# cloud=[]
# windspeed=[]
# cntr=[]
# dat=[]
#test the weather of Atlanta
# city_url=url+"&q="+"Atlanta"
# response=requests.get(city_url).json()
# pprint(response)
#response=requests.get(f"https://api.openweathermap.org/data/2.5/weather?lat={33.753746}&lon={-84.386330}&appid={weather_api_key}").json()
#pprint(response)

counter=1
Set=1
for i, city in enumerate(cities):
   if (i % 50 == 0 and i >= 50):
       Set+=1
       counter=1
   city_url=url+"&q="+city
   
   print(f"Processing Record {counter} of Set {Set} | {city} ")
   counter+=1
   # put timer here
   try:
    
       response=requests.get(city_url).json()
        
       city=response["name"]
       latitude=response['coord']['lat']
       longitude=response['coord']['lon']
       maxtemp=response["main"]["temp_max"]
       mintemp=response["main"]["temp_min"]
       hum=response["main"]["humidity"]
       cloud=response["clouds"]["all"]
       windspeed=response["wind"]["speed"]
       cntr=response["sys"]["country"]
       dat=response["dt"]
       city_data.append({
           "City":city,
           "Lat":latitude,
           "Lng":longitude,
           "Max Temp":maxtemp,
           "Humidity":hum,
           "Cloudiness":cloud,
           "Wind Speed":windspeed,
           "Country":cntr,
           "Date":dat
            
       })
   except:
       print("City not found, Skipping")
       pass
print("Data retreaval complete")

```

# Linear Regression

```python
def plot_lin(x_values,y_values,title,text_coordinates):
    (slope,intercept,rvalue,pvalue,stderr)=linregress(x_values,y_values)
    regress_values=x_values*slope+intercept
    line_eq="y="+str(round(slope,2))+"x+"+str(round(intercept,2))
    plt.scatter(x_values,y_values)
    plt.plot(x_values,regress_values,"r-")
    plt.annotate(line_eq,text_coordinates,fontsize=15,color="red")
    plt.xlabel("Latitude")
    plt.ylabel(title)
    print(f"The r-value is: {rvalue**2}")
    plt.show()
```

```python
northern=dataframe.loc[(dataframe["Lat"]>=0)]
southern=dataframe.loc[(dataframe["Lat"]<0)]
```

#### Northern Hemisphere - Max Temp vs. Latitude Linear Regression

```python
x_values=northern["Lat"]
y_values=northern["Max Temp"]
plot_lin(x_values,y_values,"Max Temp",(6,30))
```

```python
#There is weak negative correlation between Latitude and Max temp.
```

```python
x_values=southern["Lat"]
y_values=southern["Max Temp"]
plot_lin(x_values,y_values,"Max Temp",(-55,80))
```
